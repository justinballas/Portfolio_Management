```python
import pandas as pd
import requests
import datetime as dt

tickers = ['AAPL','AMZN', 'TSLA', 'GE', 'MSFT']

end = dt.date.today()
start = end - dt.timedelta(days=365*2)

def portfolio_creator(tickers, start, end):
    '''
    Gets tickers from tiingo and creates a portfolio dataframe
    '''
    
    headers = {
            'Content-Type': 'application/json',
            'Authorization': #enter tiingo api key here
            }
        
    for ticker in tickers:
        requestResponse = requests.get('https://api.tiingo.com/tiingo/daily/'
                                   + ticker +'/prices?startDate='+str(start)+'&endDate='+str(end),
                                        headers=headers)
        
        df = pd.DataFrame(requestResponse.json())
        df = df.loc[:,['date', 'adjClose']]
        df = df.rename(columns={'adjClose':ticker})  
        df.set_index('date',inplace=True)
        
        if ticker == tickers[0]:
            portfolio_df = df
        else:
            portfolio_df = pd.merge(portfolio_df,df, how='left', left_index=True, right_index=True)

    return portfolio_df

portfolio_df = portfolio_creator(tickers, start, end)
portfolio_df.to_csv('portfolio hist')
```
